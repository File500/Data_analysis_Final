```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")

df = pd.read_csv('popular_languages.csv')
df
```

```python
df = df[['Language', 'Average_Salary']]
df
```

```python
df['Average_Salary'] = df['Average_Salary'].str.replace('[$,]', '', regex=True).astype(int)
df = df.sort_values('Average_Salary', ascending=False)
df
```

```python
sns.barplot(x='Language', y='Average_Salary', data=df)
plt.xticks(rotation=90)
plt.show()
```

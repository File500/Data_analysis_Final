### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.



```python
## Write your code here
df.isna().sum()
```

```python
def impute_missing_values(df):
    
    df_imputed = df.copy()

    numeric_columns = df_imputed.select_dtypes(include=['int64', 'float64']).columns
    categorical_columns = df_imputed.select_dtypes(include=['object']).columns

    for col in numeric_columns:
        if df_imputed[col].isna().any():
            mean_value = df_imputed[col].mean()
            df_imputed[col] = df_imputed[col].fillna(mean_value)

    for col in categorical_columns:
        if df_imputed[col].isna().any():
            mode_value = df_imputed[col].mode().iloc[0]  # Get the most frequent value
            df_imputed[col] = df_imputed[col].fillna(mode_value)

    return df_imputed
```

```python
df = impute_missing_values(df)
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python

have_worked = df['LanguageHaveWorkedWith'].value_counts()
want_to_work = df['LanguageWantToWorkWith'].value_counts()

comparison_df = pd.DataFrame({
    'Have Worked With': have_worked,
    'Want to Work With': want_to_work
}).fillna(0) 


comparison_df['Total'] = comparison_df.sum(axis=1)
comparison_df = comparison_df.sort_values('Total', ascending=False)
comparison_df = comparison_df.drop('Total', axis=1)

# Take top N languages for better visualization
top_n = 15 
plot_df = comparison_df.head(top_n)

plt.figure(figsize=(12, 8))
plot_df.plot(kind='bar', width=0.8)
plt.title('Programming Languages: Current vs Desired Experience')
plt.xlabel('Programming Languages')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.legend(title='')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
## Write your code here
cross_tab = pd.crosstab(df['RemoteWork'], df['Country'])
plt.figure(figsize=(10,10))
sns.heatmap(cross_tab)
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
df['YearsCodePro'] = df['YearsCodePro'].replace({'More than 50 years':'55', 'Less than 1 year':'0'}).astype('int64')
```

```python
## Write your code here
job_exp_corr = df[['JobSat', 'YearsCodePro']].corr(method='spearman')
plt.figure(figsize=(10,10))
sns.heatmap(job_exp_corr, annot=True)
plt.show()
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
## Write your code here
cross_tab = pd.crosstab(df['EdLevel'], df['Employment'])
print("Cross-tabulation of Education Level vs Employment Status:")
print(cross_tab)

cross_tab_pct = pd.crosstab(df['EdLevel'], df['Employment'], normalize='index') * 100
print("\nPercentages (row-wise):")
print(cross_tab_pct)

ax = cross_tab_pct.plot(kind='bar', stacked=True,
                        figsize=(12, 6))

plt.title('Employment Status by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Percentage')
plt.legend(title='Employment Status', bbox_to_anchor=(1.05, 1))
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

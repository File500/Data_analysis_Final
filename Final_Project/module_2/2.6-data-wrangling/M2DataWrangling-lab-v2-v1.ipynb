### 3. Identifying and Removing Inconsistencies


<h5>3.1 Identify inconsistent or irrelevant entries in specific columns (e.g., Country).</h5>


```python
# Write your code here
object_colls = df.select_dtypes(include=['object']).columns
object_colls
```

```python
for col in object_colls:
    print(col,'--->\n',df[col].unique(),'\n','-'*100)
```

```python
df['Country'].unique() #Iran, Islamic Republic of... Venezuela, Bolivarian Republic of... Congo, Republic of the... Micronesia, Federated States of...
```

```python
df['Country'] = df['Country'].replace({'Iran, Islamic Republic of...':'Iran',
                                       'Venezuela, Bolivarian Republic of...':'Venezuela',
                                       'Congo, Republic of the...':'Congo',
                                       'Micronesia, Federated States of...':'Micronesia'})
df['Country'].unique()
```

```python
df['Country'].isna().sum()
```

<h5>3.2 Standardize entries in columns like Country or EdLevel by mapping inconsistent values to a consistent format.</h5>


```python
## Write your code here
# did in the task before for countries containing '...'
df['Country'] = df['Country'].ffill()
df['Country'].isna().sum()
```

### 4. Encoding Categorical Variables


<h5>4.1 Encode the Employment column using one-hot encoding.</h5>


```python
## Write your code here
df_encoded = pd.get_dummies(df['Employment'], prefix='Employment', drop_first=True)

df = pd.concat([df, df_encoded], axis=1)
```

### 5. Handling Missing Values


<h5>5.1 Identify columns with the highest number of missing values.</h5>


```python
## Write your code here
df.isna().sum(axis=0).sort_values(ascending=False).reset_index()
```

<h5>5.2 Impute missing values in numerical columns (e.g., `ConvertedCompYearly`) with the mean or median.</h5>


```python
object_colls = df.select_dtypes(include=['object']).columns.tolist()
numeric_cols = df.select_dtypes(include=['int64','float64']).columns.tolist()
```

```python
numeric_cols
```

```python
## Write your code here
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())
```

<h5>5.3 Impute missing values in categorical columns (e.g., `RemoteWork`) with the most frequent value.</h5>


```python
## Write your code here
df[object_colls] = df[object_colls].fillna(df[object_colls].mode().iloc[0])
```

### 6. Feature Scaling and Transformation


<h5>6.1 Apply Min-Max Scaling to normalize the `ConvertedCompYearly` column.</h5>


```python
## Write your code here
df['ConvertedCompYearly'] = df['ConvertedCompYearly'].apply(
    lambda x: ( x-df['ConvertedCompYearly'].min()) / (df['ConvertedCompYearly'].max()-df['ConvertedCompYearly'].min()))
```

<h5>6.2 Log-transform the ConvertedCompYearly column to reduce skewness.</h5>


```python
import numpy as np
```

```python
## Write your code here
df['ConvertedCompYearly'] = np.log1p(df['ConvertedCompYearly'])
```

### 7. Feature Engineering


<h5>7.1 Create a new column `ExperienceLevel` based on the `YearsCodePro` column:</h5>


```python
## Write your code here
temp_df= pd.DataFrame(columns=['ExperienceLevel'])
```

```python
df['YearsCodePro'] = df['YearsCodePro'].replace({'More than 50 years':'55', 'Less than 1 year':'0'}).astype('int64')
df['YearsCodePro'].dtype
```

```python
for entry in df['YearsCodePro'].tolist():
    
    ex_lvl = 'Beginner'
    
    if entry < 2 and entry != 0:
        ex_lvl = 'Junior'
    elif entry < 7:
        ex_lvl = 'Mid'
    elif entry >= 10:
        ex_lvl = 'Senior'
        
    temp_df.loc[len(temp_df)] = [ex_lvl]
```

```python
temp_df
```

```python
df = pd.concat([df,temp_df],  axis=1)
```
